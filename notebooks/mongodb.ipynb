```typescript
import { MongoClient } from "mongodb";
```

```typescript
const uri = "mongodb://localhost:27017";
const client = new MongoClient(uri);
const database = client.db("tradingDB");
```

```typescript
async function run() {
  try {
    const database = client.db("test");
    const movies = database.collection("movies");

    const query = { title: "Black Panter" };
    const movie = await movies.findOne(query);

    console.log(movie);
  } finally {
    await client.close();
  }
}

run().catch(console.dir);
```

```typescript
// Make new collections
(async () => {
  const createUsers = database.createCollection("users");
  const createHoldings = database.createCollection("holdings");
  const createTransactions = database.createCollection("transactions");
  Promise.all([createUsers, createHoldings, createTransactions]).then(
    (values) => {
      console.log(values);
    }
  );
})();
```
